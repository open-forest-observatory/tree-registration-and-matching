```python
from tree_registration_and_matching.example_data import (
    simulate_tree_maps,
    create_tree_crowns_from_tops,
)
import matplotlib.pyplot as plt
```

Simulate the tree top locations

```python
tree_maps = simulate_tree_maps(shift_x=5.0, shift_y=-5.0)

pred = tree_maps["pred"]
obs = tree_maps["obs"]
obs_bounds = tree_maps["obs_bounds"]
```

Generate circular tree crowns around each tree top

```python
tree_crowns = create_tree_crowns_from_tops(obs)
```

Plot the tree tops and the crowns. Note the step which ensures that the crowns are non-overlapping is more aggressive than strictly neccessary so some may have odd flat sections. 

```python
f, ax = plt.subplots(1, 1)
obs.plot(ax=ax, color="red", label="Observed Trees")
tree_crowns.boundary.plot(
    ax=ax, color="blue", linewidth=2, label="Observed Tree Crowns"
)
ax.legend()
plt.show()
```
